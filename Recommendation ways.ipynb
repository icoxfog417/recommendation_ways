# Recommendation Ways

[実践機械学習](http://www.amazon.co.jp/dp/4873116988)にて紹介されている推薦の方法をサンプルコードと共に紹介します。
具体的には、以下4つです。

1. 回帰(p153～)  
ユーザーごとの映画評価モデル(回帰モデル)を作成し、それで予測を行います。具体的には、実際にユーザーが行った評価を、他のユーザーの評価から予測します(似ているユーザーの評価には高い重みがつくはず)。
2. 協調フィルタリング(p160～)  
映画に対して評価を行っているかどうかの二値(0/1)により似ているユーザーを見つけ、似ているユーザーの評価から予測します。具体的には、似ているかどうかを相関係数から判断し、類似ユーザーのトップ100の平均評価を予測値とします。
3. 類似映画によるレコメンド(p162～)  
2の手法をユーザー同士でなく映画同士に適用し、似ている映画の評価から予測します。
4. バスケット分析(p165～)  
一緒に買われる可能性の高い集合を見つけて、ある商品が買われたときに一緒に買われる可能性が高いものを提案します。具体的には、全商品の集合について実際の購買データ中に発生した個数を数え上げ、カウントの高いものを列挙していきます。
5. アンサンプル学習(p163)  
上記の手法を組み合わせて予測精度を上げる方法です。

## 協調フィルタリング

実践機械学習では二値による判定を行っていますが、ここではオーソドックスにレビューの評価値を使用し類似ユーザーを見つけ、そこからレコメンドを行ってみます。使用するデータは観光地の評価データで、データフォーマットは以下のようになっています。

* ユーザーID
* 観光地ID
* 評価値

実装に当たり、主要な関数は以下2つです。

* calc_collaborative_users: 与えられたユーザーと他のユーザーの近さを返す。「近さ」の計算には相関係数を使用。
* filter_by_collaborative_users: `calc_collaborative_users`で計算したユーザーとその近さを元に、推薦する観光地を返す。具体的には与えられたユーザーに近いN人のユーザーを利用し、「近さ」を重みとして観光地に対する評価の加重平均を計算し、その結果を返す。



```python
import numpy as np
from data_loader import load
import collaborative_filtering as cf


# データのロード
user_reviews = load()

# 推薦を行うユーザーIDを設定
target_user_id = user_reviews.users[0]

# target_user_idへの近さを計算
collaborators = cf.calc_collaborative_users(target_user_id, user_reviews)

# 計算した「近さ」を利用し、近いユーザー(トップNユーザー)の評価を加重平均で計算する。
N = 10
places = cf.filter_by_collaborative_users(user_reviews, collaborators, to_rank=N)


# 指定ユーザーに近いユーザーを表示
display_limit = 5
print("user {0}'s collaborators are below (top5)".format(target_user_id))
for c in collaborators[:display_limit ]:
    user_id = user_reviews.users[c[0]]
    print("{0}: {1}".format(user_id, c[1]))


# トップNユーザーの加重平均)が高い順に表示
print("favorite places will be below")
for p in places[:display_limit ]:
    place_id = user_reviews.places[p[0]]
    print("{0}: {1}".format(place_id, p[1]))
```
